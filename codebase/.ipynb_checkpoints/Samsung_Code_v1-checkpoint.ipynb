## (5-1) 통계 분석 - Kolmogorov-Smirnov one sample Test (정규성 검증) : chamber_stat

```python
#####ks가 되도록
chamber_stat = pd.concat([chamber_perf['VALUE', 'count_nonzero'], chamber_perf['VALUE', 'mean']], axis=1, keys=['count', 'mean'])

user_defined_alpha_shapiro = 0.05

def eval_normality(unique_Chamber):
    chamber = raw_4_wafer[raw_4_wafer['S_E_CHAMBER_ID'] == unique_Chamber]
    if len(chamber['VALUE']) > 3:
        d, p_value = stats.shapiro(chamber['VALUE'])
    else:
        p_value = 0.0
    return p_value

chamber_stat['shapiro_p_value'] = chamber_stat.index.to_series().apply(eval_normality)
chamber_stat['Normality'] = chamber_stat['shapiro_p_value'] > user_defined_alpha_shapiro 
chamber_stat
```

## (5-2) 통계 분석 - ANOVA : chamber_stat

```python
#ANOVA

#step별 정리
unique_Steps = raw_4_wafer['STEP_SEQ'].unique()
user_defined_alpha_ANOVA = 0.05

for unique_Step in unique_Steps:
    #raw_4_wafer를 특정 STEP으로 filtering
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    
    unique_chambers = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]['S_E_CHAMBER_ID'].unique()

    #step 내 설비들의 VALUE를 저장
    step_chamber_values = {}
    for unique_chamber in unique_chambers:
        #filtered['S_E_CHAMBER_ID']가 필요, 모든 observation이 필요하므로
        step_chamber_values[unique_chamber]=(filtered.loc[filtered['S_E_CHAMBER_ID']==unique_chamber, 'VALUE'])

    statistic, p_value = stats.f_oneway(*step_chamber_values.values())


    chamber_stat.loc[unique_chambers, 'anova_p_value'] = p_value

chamber_stat['ANOVA'] = chamber_stat['anova_p_value'] < user_defined_alpha_ANOVA
chamber_stat
```

## (5-3) 통계 분석 - Kruskal Wallis H-Test : chamber_stat

```python
#KRUSKAL
user_defined_alpha_KRUSKAL =0.05
#step별 정리
unique_Steps = raw_4_wafer['STEP_SEQ'].unique()
for unique_Step in unique_Steps:
    #raw_4_wafer를 특정 STEP으로 filtering
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    
    unique_chambers = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]['S_E_CHAMBER_ID'].unique()

    #step 내 설비들의 VALUE를 저장
    step_chamber_values = {}
    for unique_chamber in unique_chambers:
        #filtered['S_E_CHAMBER_ID']가 필요, 모든 observation이 필요하므로 
        step_chamber_values[unique_chamber]=(filtered.loc[filtered['S_E_CHAMBER_ID']==unique_chamber, 'VALUE'])

    statistic, p_value = stats.kruskal(*step_chamber_values.values())


    chamber_stat.loc[unique_chambers, 'kruskal_p_value'] = p_value
chamber_stat['KRUSKAL'] = chamber_stat['kruskal_p_value'] < user_defined_alpha_KRUSKAL
chamber_stat
```

## (5-4) 통계 분석 - Linear Contrast : chamber_stat_LC

```python
def Linear_Contrast(row, other_rows, unique_Step):
    #N is total observations
    N = sum(other_rows['VALUE', 'count_nonzero'])
    
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    filtered_VALUE = filtered.loc[filtered['S_E_CHAMBER_ID'].isin(other_rows.index), 'VALUE']
    
    #calculate SS_error
    sum_y_squared = sum([value**2 for value in filtered_VALUE])
    SS_error = sum_y_squared - sum(filtered.groupby('S_E_CHAMBER_ID').sum()['VALUE']**2/filtered.groupby('S_E_CHAMBER_ID').size())
    
    #exclude 비교대상
    other_rows = other_rows.drop(row.name)
    
    #k is # groups
    k = len(other_rows.index)
    
    #calcualte SS contrast
    L = k*row['VALUE', 'mean']
    for mean in other_rows['VALUE', 'mean']:
        L -= mean
    
    denominator = k*k/row['VALUE', 'count_nonzero']
    for no_obs in other_rows['VALUE', 'count_nonzero']:
        denominator += 1/no_obs
        N +=no_obs
        
    #degrees of freedom
    DFbetween = 1
    DFwithin = N - k
    
    MS_error = SS_error / DFwithin
    
    std = math.sqrt(MS_error*denominator)
    
    t_contrast = abs(L)/std
    
    p_value = stats.t.sf(t_contrast, DFwithin)
    return p_value

def Nonparametric_Linear_Contrast(row, other_rows, unique_Step):
    #N is total observations
    other_rows = other_rows.drop(row.name)
    
    
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    other_rows_filtered_VALUE = filtered.loc[filtered['S_E_CHAMBER_ID'].isin(other_rows.index), 'VALUE']
    #print(other_rows_filtered_VALUE)
    row_filtered_VALUE = filtered.loc[filtered['S_E_CHAMBER_ID'] == row.name, 'VALUE']
    #print(row_filtered_VALUE)
    statistic, p_value = stats.mannwhitneyu(other_rows_filtered_VALUE, row_filtered_VALUE)
    #print(p_value)
    return p_value
```

```python
#통계분석
#기본적으로 DUMMY화 시켜놓고, BOB/WOW 설비만 재설정

chamber_stat_LC = chamber_stat.copy()
user_defined_alpha_LC = 0.01
user_defined_alpha_NLC = 0.01

chamber_stat_LC['LC_p_value'] = 1.0
chamber_stat_LC['LCresult'] = 'DUMMY'

unique_Steps = raw_4_wafer['STEP_SEQ'].unique()
#count_k=0
#count_a=0
#count = 0
for unique_Step in unique_Steps:
    #count +=1
    #raw_4_wafer를 특정 STEP으로 filtering
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    #해당 step의 chamber_stat, filtered['S_E_CHAMBER_ID'].unique()가 필요, 어떤 chamber가 있는지가 필요하므로
    step_chamber_stat = chamber_stat_LC.loc[filtered['S_E_CHAMBER_ID'].unique()]
    
    #pass Normality test
    if all(step_chamber_stat['shapiro_p_value'] > user_defined_alpha_shapiro):
        #count_a+=1
        #ANOVA result: 유의미한 차이 없음
        if (step_chamber_stat['anova_p_value'] > user_defined_alpha_ANOVA).any():
            #chamber_stat.loc[step_chamber_stat.index]['result'] = 'DUMMY' #왜 안되냐?
            chamber_stat_LC['LCresult'][step_chamber_stat.index] = 'DUMMY'
            #print(chamber_stat.loc[step_chamber_stat.index]['result'])
        else:
            #Linear Contrast 수행
            sorted_chamber_perf = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].sort_values(by=[('VALUE','mean')],ascending=False)
            #print(sorted_chamber_perf)
            for index, row in sorted_chamber_perf.iterrows():
                p_value = Linear_Contrast(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step)
                chamber_stat_LC['LC_p_value'][row.name]  = p_value
                ###Linear Contrast 결과 유의미한 경우
                if p_value < user_defined_alpha_LC:
                    chamber_stat_LC['LCresult'][row.name] = 'BOB'
                else:
                    break
            sorted_chamber_perf = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].sort_values(by=[('VALUE','mean')])            
            for index, row in sorted_chamber_perf.iterrows():
                p_value = Linear_Contrast(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step)
                chamber_stat_LC['LC_p_value'][row.name]  = p_value
                if p_value < user_defined_alpha_LC:
                    chamber_stat_LC['LCresult'][row.name] = 'WOW'
                else:
                    break
                    
    #Non-parametric method
    else:
        #count_k+=1
        if (step_chamber_stat['kruskal_p_value'] > user_defined_alpha_KRUSKAL).any():
            chamber_stat_LC['LCresult'][step_chamber_stat.index] = 'DUMMY'
            
        else:
            ###non-parametric linear contrast
            
            #from the best
            sorted_chamber_perf = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].sort_values(by=[('VALUE','mean')],ascending=False)
            for index, row in sorted_chamber_perf.iterrows():
                p_value = Nonparametric_Linear_Contrast(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step)
                chamber_stat_LC['LC_p_value'][row.name]  = p_value
                ###Linear Contrast 결과 유의미한 경우
                if p_value < user_defined_alpha_NLC:
                    chamber_stat_LC['LCresult'][row.name] = 'BOB'
                else:
                    chamber_stat_LC['LCresult'][row.name] = 'DUMMY'
                    break
            #from the worst
            sorted_chamber_perf = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].sort_values(by=[('VALUE','mean')])            
            for index, row in sorted_chamber_perf.iterrows():
                p_value = Nonparametric_Linear_Contrast(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step)
                chamber_stat_LC['LC_p_value'][row.name]  = p_value
                if p_value < user_defined_alpha_NLC:
                    chamber_stat_LC['LCresult'][row.name] = 'WOW'
                else:
                    chamber_stat_LC['LCresult'][row.name] = 'DUMMY'
                    break
    
#WOW 확인
#chamber_stat_LC.loc[chamber_stat_LC['LCresult']=='WOW']  
chamber_stat_LC
                        
                        
                        
```

## (5-) 통계 분석 - Multiple T-Test

```python
def Multiple_t_test(row, other_rows, unique_Step, label):
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    count = len(other_rows.index.unique())
    other_rows = other_rows.drop(row.name)
    ref = filtered.loc[filtered['S_E_CHAMBER_ID'] == row.name, 'VALUE']
    for index, instance in other_rows.iterrows():
        
        target = filtered.loc[filtered['S_E_CHAMBER_ID'] == instance.name, 'VALUE']
        t_statistic, p_value = stats.ttest_ind(ref, target)
        #Bonferonni
        #if p_value < user_defined_alpha_tt/count:
        if p_value < user_defined_alpha_tt:
            if chamber_stat_MTTest.loc[chamber_stat_MTTest['S_E_CHAMBER_ID'] == row.name, 'mean'] > chamber_stat_MTTest.loc[chamber_stat_MTTest['S_E_CHAMBER_ID'] == instance.name, ['VALUE', 'mean']]:
                BOB.append(chamber_stat_MTTest['t_test_result'][row.name])
                WOW.append(chamber_stat_MTTest['t_test_result'][instance.name])
            else:
                BOB.append(chamber_stat_MTTest['t_test_result'][instance.name])
                WOW.append(chamber_stat_MTTest['t_test_result'][row.name])

def Multiple_u_test(row, other_rows, unique_Step, label):
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
    count = len(other_rows.index.unique())
    other_rows = other_rows.drop(row.name)
    ref = filtered.loc[filtered['S_E_CHAMBER_ID'] == row.name, 'VALUE']
    for index, instance in other_rows.iterrows():
        target = filtered.loc[filtered['S_E_CHAMBER_ID'] == instance.name, 'VALUE']
        u_statistic, p_value = stats.mannwhitneyu(ref, target)
        #if p_value < user_defined_alpha_ut/count:
        if p_value < user_defined_alpha_ut:
            if chamber_stat_MTTest.loc[row.name]['mean'] > chamber_stat_MTTest.loc[instance.name]['mean']:
                BOB.append(row.name)
                WOW.append(instance.name)
            else:
                BOB.append(instance.name)
                WOW.append(row.name)
                
```

```python
#Multiple t-test
user_defined_alpha_tt = 0.01
user_defined_alpha_ut = 0.01
user_defined_tt_BOB_criterion = 8
user_defined_tt_WOW_criterion = 7
chamber_stat_MTTest = chamber_stat.copy()




chamber_stat_MTTest['t_test_result'] = 'DUMMY'

unique_Steps = raw_4_wafer['STEP_SEQ'].unique()
#print (unique_Steps)
#a_count = 0
k_count = 0
#count = 0
count_z=0
for unique_Step in unique_Steps:
    BOB = []
    WOW = []
    #count+=1
    #raw_4_wafer를 특정 STEP으로 filtering
    filtered = raw_4_wafer[raw_4_wafer['STEP_SEQ'] == unique_Step]
#    print(filtered)
    #해당 step의 chamber_stat, filtered['S_E_CHAMBER_ID'].unique()가 필요, 어떤 chamber가 있는지가 필요하므로
    step_chamber_stat2 = chamber_stat_MTTest.loc[filtered['S_E_CHAMBER_ID'].unique()]
    
    #pass Normality test
    if all(step_chamber_stat2['shapiro_p_value'] > user_defined_alpha_shapiro):
        #a_count +=1
        #ANOVA result: 유의미한 차이 없음
        if (step_chamber_stat2['anova_p_value'] > user_defined_alpha_ANOVA).any():
            #chamber_stat.loc[step_chamber_stat.index]['result'] = 'DUMMY' #왜 안되냐?
            chamber_stat_MTTest['t_test_result'][step_chamber_stat2.index] = 'DUMMY'
            #print(chamber_stat.loc[step_chamber_stat.index]['result'])
            
        else:
            #Multiple t-test 수행
            #Linear contrast와 마찬가지로 상위 20% 하위 20% 설비에 대해서만(환경변수) 타 설비와 t-test를 진행 
            VALUE = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()]['VALUE', 'mean']
            for index, row in chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].iterrows():
                
                if row['VALUE', 'mean'] > BOB_VALUE:
                    label = 'BOB'
                    Multiple_t_test(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step, label)
                    
                elif row['VALUE', 'mean'] <WOW_VALUE:
                    label = 'WOW'
                    Multiple_t_test(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step, label)
    else:
        #k_count +=1
        if (step_chamber_stat2['kruskal_p_value'] > user_defined_alpha_KRUSKAL).any():
            chamber_stat_MTTest['t_test_result'][step_chamber_stat2.index] = 'DUMMY'
        else:
            #nonparametric method: Mann-Whitney U-test

            VALUE = chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()]['VALUE', 'mean']
            """
            V_BOB_RELA = 80
            V_WOW_RELA = 20
            BOB_VALUE = np.percentile(VALUE, V_BOB_RELA)
            WOW_VALUE = np.percentile(VALUE, V_WOW_RELA)
            ###CAUTION
            #특정 STEP 내 챔버의 수율 평균의 percentile을 return함
            ###print (BOB_VALUE, WOW_VALUE)
            """
            for index, row in chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()].iterrows():
                if row['VALUE', 'mean'] > BOB_VALUE:
                    label = 'BOB'
                    Multiple_u_test(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step, label)
                    
                elif row['VALUE', 'mean'] <WOW_VALUE:
                    label = 'WOW'
                    Multiple_u_test(row, chamber_perf.loc[filtered['S_E_CHAMBER_ID'].unique()], unique_Step, label)
    #l3 = [x for x in l1 if x not in l2]
    BOB = list(set([chamber for chamber in BOB if chamber not in WOW]))
    BOB = [chamber for chamber in BOB if chamber_stat_MTTest.loc[chamber]['mean'] > user_defined_tt_BOB_criterion]
    WOW = list(set([chamber for chamber in WOW if chamber not in BOB]))
    WOW = [chamber for chamber in WOW if chamber_stat_MTTest.loc[chamber]['mean'] < user_defined_tt_WOW_criterion]
    chamber_stat_MTTest['t_test_result'][BOB] = 'BOB'
    chamber_stat_MTTest['t_test_result'][WOW] = 'WOW'
chamber_stat_MTTest    
#chamber_stat_MTTest.loc[chamber_stat_MTTest['t_test_result'] == 'WOW']
                

```

## (6) 사용자 정의 수율 분석 (수율 평균 - BOB/WOW: 8.0/7.0, BOB/WOW 비율 - 20%)

```python
chamber_perf_user_defined = chamber_perf.copy()

user_defined_BOB_avg = 8
user_defined_WOW_avg = 7

user_defined_BOB_Prop = 0.20
user_defined_WOW_Prop = 0.20

user_defined_chamber_weight = 0.5
user_defined_step_weight = 0.5


chamber_perf_user_defined['AVG_Test'] = np.where(chamber_perf_user_defined['VALUE', 'mean']  > user_defined_BOB_avg, 'BOB', np.where(chamber_perf_user_defined['VALUE', 'mean'] < user_defined_WOW_avg, 'WOW', 'DUMMY'))
chamber_perf_user_defined['Prop_Test'] = np.where(user_defined_chamber_weight*chamber_perf_user_defined['BOB', 'mean'] + user_defined_step_weight*chamber_perf_user_defined['BOB/step'] > user_defined_BOB_Prop, np.where(user_defined_chamber_weight*chamber_perf_user_defined['WOW', 'mean'] + user_defined_step_weight*chamber_perf_user_defined['WOW/step'] > user_defined_WOW_Prop, 'DUMMY', 'BOB'), np.where(user_defined_chamber_weight*chamber_perf_user_defined['WOW', 'mean'] + user_defined_step_weight*chamber_perf_user_defined['WOW/step'] > user_defined_WOW_Prop, 'WOW', 'DUMMY'))
chamber_perf_user_defined[['AVG_Test', 'Prop_Test']]
```

```python
mode = 'and'
mode = 'or'

def and_assign(chamber):
    x = chamber_perf_user_defined.loc[chamber]['AVG_Test']
    y = chamber_perf_user_defined.loc[chamber]['Prop_Test']
    x = x.values
    y = y.values
    if x==y:
        return x[0]
    else:
        return 'DUMMY'

def or_assign(chamber):
    x = chamber_perf_user_defined.loc[chamber]['AVG_Test']
    y = chamber_perf_user_defined.loc[chamber]['Prop_Test']
    x = x.values
    y = y.values
    x = x[0]
    y = y[0]
    if x==y:
        return x
    else:
        if x != 'DUMMY':
            return x
        elif y != 'DUMMY':
            return y
        else:
            return 'DUMMY'
        

if mode == 'and':
    chamber_perf_user_defined['UDresult'] = chamber_perf_user_defined.index.to_series().apply(and_assign)
elif mode == 'or':
    chamber_perf_user_defined['UDresult'] = chamber_perf_user_defined.index.to_series().apply(or_assign)
chamber_perf_user_defined
```
